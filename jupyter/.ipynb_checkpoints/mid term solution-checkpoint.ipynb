#### 14.	ให้แสดงค่าเฉลี่ย (และ s.d.) ของราคาขายสินค้าของแต่ละร้านค้า โดยแสดงเพียงร้านที่มีค่าเฉลี่ยสูงสุด 10 อันดับและต่ำสุด 10 อันดับ (4 คะแนน)

```python
tmp = df[['shop_id', 'item_price']]
price_by_shop = tmp.groupby(tmp['shop_id'])

shop_mean = price_by_shop.mean().reset_index()
shop_mean['mean'] = shop_mean.pop('item_price')

shop_sd = price_by_shop.agg(np.std).reset_index()
shop_sd['S.D.']   = shop_sd.pop('item_price')

output = pd.merge(shop_mean, shop_sd, on=['shop_id'])
output = output.sort_values(by=['mean'], ascending=False)
output.head(10)
```

```python
output.tail(10)
```

#### 15.	ให้นำยอดขายของทุกร้านมาเรียงจากมากไปน้อย แล้วแบ่งจัดกลุ่มร้านทั้งหมดออกเป็น 6 กลุ่ม (ตามจำนวนร้านโดยปัดเศษตามสมควร) จากนั้นให้แสดงยอดขายรวมของกลุ่มทั้ง 6 กลุ่มนั้นเปรียบเทียบกัน (เป็นร้อยละหรือสัดส่วนอย่างสวยงาม) (6 คะแนน)

```python
# Credit: adjusted from Grp 04-13-54
#
shop_price = df[['shop_id', 'item_price']]
shop_price = shop_price.sort_values(by=['item_price'], ascending=False)
shop_price = shop_price.groupby(shop_price['shop_id']).sum()
six_groups = pd.qcut(shop_price['item_price'], 6).reset_index()
six_groups['Range'] = six_groups.pop('item_price')

shop_price = shop_price.reset_index()
shop_price['Price'] = shop_price.pop('item_price')
shop_price = pd.merge(shop_price, six_groups, on=['shop_id'])
shop_price = shop_price.groupby(shop_price['Range'])
shop_price = shop_price['Price'].sum()
shop_price
```

```python
lb = 'G1', 'G2', 'G3', 'G4','G5','G6'
cl = ['gold', 'yellowgreen', 'lightskyblue', 'sandybrown', 'aquamarine','salmon']
explode = (0, 0, 0, 0, 0, 0.1)  # explode 1st slice

plt.figure(figsize=(15,8))
plt.pie(shop_price, explode=explode, labels=lb, colors=cl, 
        autopct='%1.1f%%', shadow=True, startangle=90,
        textprops={'fontsize': 16})
plt.axis('equal')
plt.show
```

#### 17.	ให้แสดงจำนวนสินค้าที่ขายได้รวมในแต่ละวันของสัปดาห์ และ ยอดขายรวมในแต่ละวันของสัปดาห์ โดยนำเสนอข้อมูลทั้งสองนี้ข้าง ๆ กัน (6 คะแนน)

```python
# remove some columns that are likely not to be used anymore
df = df.drop(columns=['date_block_num', 'item_name', 'item_category_id', 'dof'])

# กำหนดคอลัม Week เพื่อบอกวันเป็นชื่อวัน เช่น วันจันทร์ อังคาร

df['dow'] = df['date'].dt.day_name()
df2013 = df.loc[df['year'] == 2013]
df2014 = df.loc[df['year'] == 2014]
df2015 = df.loc[df['year'] == 2015]
df2013.head()
```

```python
def show_chart_of_items_and_amount(data, year):
    fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(20,6))

    pv_count = pd.pivot_table(data, 
                   index=['year', 'dow'],
                   values=['item_cnt_day'], 
                   aggfunc=np.sum,fill_value=0).plot(kind='bar', ax=axes[0], color='blue', alpha=0.5)

    pv_amount = pd.pivot_table(data, 
                   index=['year', 'dow'],
                   values=['amount'], 
                   aggfunc=np.sum,fill_value=0).plot(kind='bar', ax=axes[1], color='green', alpha=0.5)

    axes[0].set_title('Number of items sold in ' + str(year));
    axes[1].set_title('Amount sold in ' + str(year));
    axes[0].set_xlabel('Day of Week');
    axes[0].set_ylabel('units');
    axes[1].set_xlabel('Day of Week');
    axes[1].set_ylabel('baht');
```

```python
show_chart_of_items_and_amount(df2013, 2013)
```

```python
show_chart_of_items_and_amount(df2014, 2014)
```

```python
show_chart_of_items_and_amount(df2015, 2015)
```
